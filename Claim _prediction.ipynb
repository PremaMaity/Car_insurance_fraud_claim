```python
df['cylinder']=df['cylinder'].astype('object')
df['ncap_rating']=df['ncap_rating'].astype('object')
df['make']=df['make'].astype('object')
df['make']=df['make'].astype('object')
df['airbags']=df['airbags'].astype('object')
df['displacement']=df['displacement'].astype('object')
df['is_claim']=df['is_claim'].astype('object')
df['gear_box']=df['gear_box'].astype('object')
```

```python
categorical_col=[features for features in df.columns if df[features].dtype=='O']
print('The  number of categorical features are: ',len(categorical_col))
print(categorical_col)
```

```python
Numerical_col=[features for features in df.columns if df[features].dtype!='O']
print('The  number of numerical features are: ',len(Numerical_col))
print(Numerical_col)
```
